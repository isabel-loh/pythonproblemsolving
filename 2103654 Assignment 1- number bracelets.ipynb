First create functions to compute bracelets.

```python
#create function to perform addition mod 10
def add_mod_10(a, b):
    return (a + b) % 10

#function to generate bracelet
def generate_bracelet(start_a, start_b, max_length=100):
    a, b = start_a, start_b
    bracelet = [a, b]

    while len(bracelet) < max_length:
        c = add_mod_10(a, b)
        bracelet.append(c)
        a, b = b, c

    return bracelet

# Example: Start with 1 and 5
example = generate_bracelet(1,5)

print("Bracelet starting with (1,5):", example)
```

Write a function to look for when bracelets loop:

```python
#finds bracelet without repeats

def find_repeated_pairs(sequence):
    seen_pairs = set()
    current_pair = (sequence[0], sequence[1])

    for i in range(2, len(sequence)):
        if current_pair in seen_pairs:
            print(f"Repeated pair found: {current_pair}")

            # Append the sequence to eliminate repeated pairs
            new_sequence = sequence[:i-2]
            print(f"Bracelet without repeats: {new_sequence}")
            print("The bracelet has period", len(new_sequence))
            return new_sequence

        seen_pairs.add(current_pair)
        new_number = add_mod_10(current_pair[0], current_pair[1])
        current_pair = (current_pair[1], new_number)

    print("No repeated pair found in the given sequence.")
    
    return sequence

# Call the function to find repeated pairs in the sequence
new_sequence = find_repeated_pairs(example)
print('The next two numbers would be 7+4=11 which is 1 mod 10, and then 4+1=5 which is the same pair we started with.')
```
