### Write a function to run the hat numbers experiment

```python
import random
from math import gcd

# Function to perform the experiment
def final_piece(num):
    # Create a list of numbers from 1 to num_pieces
    pieces = list(range(1, num + 1))
    
    # Repeat the process until there is only one piece left
    while len(pieces) > 1:
        # Draw two random pieces
        piece1, piece2 = random.sample(pieces, 2)
        # Calculate the difference and put it back in the hat
        diff = abs(piece1 - piece2)
        pieces.remove(piece1)
        pieces.remove(piece2)
        pieces.append(diff)
    
    # Return the final piece
    return pieces

# Number of pieces in the hat
num = 2024

# Perform the experiment
final = final_piece(num)

print("The final piece of paper is:", final[0])
```

## Perform the experiment multiple times and create graphics

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Perform the experiment multiple times
results = []
results2 = []
results3 = []

# generate hat numbers
for i in range(1000):
    results.append(final_piece(num)[0])
    results2.append(final_piece(num)[0])
    results3.append(final_piece(num)[0])

# combine results and create a corresponding group array
all_results = results + results2 + results3
groups = ['Results 1'] * len(results) + ['Results 2'] * len(results2) + ['Results 3'] * len(results3)

# colors for each violin plot
palette = {'Results 1': 'skyblue', 'Results 2': 'salmon', 'Results 3': 'lightgreen'}

# violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(x=groups, y=all_results, inner="point", cut=0, density_norm="count", palette=palette)
plt.title('Violin Plot of Results',fontsize=16)
plt.ylabel('Final piece number',fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.grid(True)
plt.savefig('violin.png')

#boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x=results,color='skyblue')
plt.title("Distribution of Final Pieces",fontsize=16)
plt.xlabel("Final piece value",fontsize=16)
plt.ylabel("Frequency",fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.savefig('boxplot.png')

#histogram
plt.figure(figsize=(10, 6))
plt.hist(results, bins=100, color='skyblue', edgecolor='black')
plt.title('Histogram',fontsize=16)
plt.xlabel('Final piece value',fontsize=16)
plt.ylabel('Frequency',fontsize=16)
plt.grid(True)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.savefig('histogram.png')
```
